```python
import json

textfile = "/home/dnlab/Data-B/5G_LLM/5G_data/gold_5G_v4.txt"

def process_text(text):
    # Split the text based on two line breaks to separate instructions and outputs
    sections = text.split('\n\n\n')
    instruction_output_list = []

    for section in sections:
        # Split each section into lines
        lines = section.split('\n')
        # The first line is the instruction
        instruction = lines[0]
        # The remaining lines are the output
        output = '\n'.join(lines[1:])
        # Create the instruction-output dictionary
        instruction_output_dict = {
            "instruction": instruction,
            "output": output
        }
        # Append to the list
        instruction_output_list.append(instruction_output_dict)
    
    return instruction_output_list

# Open and read the txt file
with open(textfile, 'r', encoding='utf-8') as file:
    text = file.read()

# Process the text to extract instruction-output pairs
instruction_output_list = process_text(text)

# Save the list to a JSON file
with open('5G_spec_knowledge.json', 'w', encoding='utf-8') as file:
    json.dump(instruction_output_list, file, indent=4)
```

## Evaluate the performance of a 5G network infrastructure using various Key Performance Indicators (KPIs)

`Total Network Capacity`: This KPI measures the combined capacity of the spectrum and small cells, which indicates the overall bandwidth available in the network. This is fundamental to understanding whether the network can handle the current and projected user demand.

`Capacity per Area`: This KPI evaluates how the total capacity is distributed over a given area (in Mbps per square kilometer). It's useful for assessing network coverage and identifying areas that may need additional infrastructure to improve service.

`Capacity per Point`: This measures the capacity available per network point (like a cell tower), giving insight into how well the network can serve users at specific points and potentially highlighting where capacity is strained.

`Cost per Capacity`: By dividing the total cost by the total capacity, this KPI assesses the cost-effectiveness of the network capacity. It helps in financial planning and can guide investment decisions to optimize the cost-to-capacity ratio.

`Cost per Area`: This KPI multiplies the cost per square kilometer by the area size to determine the cost of network coverage per area. It's a critical measure for understanding the investment required to maintain or expand network coverage.

`Surplus per Area`: This KPI reveals the excess capacity available per area, which is important for planning future network expansions or scaling back if there's significant unused capacity.

Input: 5G Mobile Broadband data from '5G_Infrastructure/demand_driven_postcode_data_results.csv'.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the reduced CSV file into a pandas DataFrame
df = pd.read_csv('5G_Infrastructure/demand_driven_postcode_data_results.csv')

# Categorize columns based on specific keywords
spectrum_columns = [col for col in df.columns if 'spectrum.capacity' in col.lower()]
small_cells_columns = [col for col in df.columns if 'small.cells.capacity' in col.lower()]
num_points_columns = [col for col in df.columns if 'numpoints' in col.lower()]

# Define a function to calculate per user capacity increase due to 5G spectrum bands
def calculate_per_user_capacity_increase(df):
    # Calculate per user capacity for LTE using spectrum capacity
    df['per_user_capacity_LTE'] = df[spectrum_columns].sum(axis=1) / df[num_points_columns].sum(axis=1)

    # Calculate per user capacity for 5G using spectrum and small cells capacity
    df['per_user_capacity_5G'] = (df[spectrum_columns].sum(axis=1) + df[small_cells_columns].sum(axis=1)) / df[num_points_columns].sum(axis=1)
    
    # Calculate the increase in per user capacity
    df['per_user_capacity_increase'] = df['per_user_capacity_5G'] - df['per_user_capacity_LTE']
    
    return df

# Apply the function to calculate per user capacity increase
df = calculate_per_user_capacity_increase(df)

# Plot the per user capacity increase and mark the threshold where small cells are needed
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['per_user_capacity_increase'], label='Per User Capacity Increase')
plt.axhline(y=5000, color='r', linestyle='--', label='Threshold for Small Cells')  # Adjust the threshold value as needed

# Label the plot
plt.title('Per User Capacity Increase with 5G Spectrum Bands')
plt.xlabel('Index')
plt.ylabel('Capacity Increase (Mbps)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('5G_Infrastructure/demand_driven_postcode_data_results.csv')

# Automatically detect relevant columns by keyword
cost_columns = [col for col in df.columns if 'cost' in col.lower()]
ran_macro_columns = [col for col in df.columns if 'ran.macro' in col.lower()]
civil_works_macro_columns = [col for col in df.columns if 'civil.works.macro' in col.lower()]
ran_small_cells_columns = [col for col in df.columns if 'ran.small.cells' in col.lower()]
civil_works_small_columns = [col for col in df.columns if 'civil.works.small' in col.lower()]
strategy_column = [col for col in df.columns if 'strategy' in col.lower()][0]

# Ensure we have at least one column for each category
assert ran_macro_columns and civil_works_macro_columns and ran_small_cells_columns and civil_works_small_columns and strategy_column, "Required columns not found."

# Replace NaN values and zeros in cost columns to avoid division by zero
df[cost_columns] = df[cost_columns].replace({0: np.nan}).fillna(method='bfill').fillna(method='ffill').fillna(0)

# Calculate cost percentages directly within the DataFrame using .loc to avoid SettingWithCopyWarning
df['RAN_cost_percentage'] = 0
df['Civil_works_cost_percentage'] = 0

# Calculate percentages for each strategy
df.loc[df[strategy_column] == 'Spectrum Integration', 'RAN_cost_percentage'] = df[ran_macro_columns].sum(axis=1) / df[cost_columns].sum(axis=1) * 100
df.loc[df[strategy_column] == 'Spectrum Integration', 'Civil_works_cost_percentage'] = df[civil_works_macro_columns].sum(axis=1) / df[cost_columns].sum(axis=1) * 100

df.loc[df[strategy_column] == 'Small Cells', 'RAN_cost_percentage'] = df[ran_small_cells_columns].sum(axis=1) / df[cost_columns].sum(axis=1) * 100
df.loc[df[strategy_column] == 'Small Cells', 'Civil_works_cost_percentage'] = df[civil_works_small_columns].sum(axis=1) / df[cost_columns].sum(axis=1) * 100

df.loc[df[strategy_column] == 'Hybrid', 'RAN_cost_percentage'] = (df[ran_macro_columns + ran_small_cells_columns].sum(axis=1)) / df[cost_columns].sum(axis=1) * 100
df.loc[df[strategy_column] == 'Hybrid', 'Civil_works_cost_percentage'] = (df[civil_works_macro_columns + civil_works_small_columns].sum(axis=1)) / df[cost_columns].sum(axis=1) * 100

# Group by strategy and calculate the mean cost percentages for RAN and Civil Works
grouped_costs = df.groupby(strategy_column).mean()[['RAN_cost_percentage', 'Civil_works_cost_percentage']]

# Plot the cost distribution for each strategy
grouped_costs.plot(kind='bar', stacked=False, figsize=(12, 7), title='Cost Distribution by Strategy')
plt.ylabel('Percentage of Total Cost')
plt.xlabel('Strategy')
plt.legend(title='Cost Components')
plt.show()

# Print the grouped costs for verification
print(grouped_costs)
```

##  Maximum average capacity per user traffic level 

Instruction: Generate Python code to calculate and chart the average traffic per user for LTE and 5G networks across different geographic types. The script must avoid division by zero errors and provide a comparative visualization that showcases the average user traffic for both LTE and 5G by geotype.

Input: CSV dataset '5G_Infrastructure/demand_driven_postcode_data_results.csv', which includes the number of connection points and the capacity for LTE and 5G small cells.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('5G_Infrastructure/demand_driven_postcode_data_results.csv')

# Detect relevant columns by keywords
spectrum_capacity_columns = [col for col in df.columns if 'spectrum.capacity' in col.lower()]
small_cells_capacity_columns = [col for col in df.columns if 'small.cells.capacity' in col.lower()]
num_points_columns = [col for col in df.columns if 'numpoints' in col.lower()]
geotype_columns = [col for col in df.columns if 'geotype' in col.lower()]

# Ensure we have at least one column for each category
assert spectrum_capacity_columns and small_cells_capacity_columns and num_points_columns and geotype_columns, "Required columns not found."

# Select the first column from each category
spectrum_capacity_column = spectrum_capacity_columns[0]
small_cells_capacity_column = small_cells_capacity_columns[0]
num_points_column = num_points_columns[0]
geotype_column = geotype_columns[0]

# Remove rows with no connection points to avoid division by zero
df = df[df[num_points_column] > 0]

# Calculate per user traffic level for LTE and 5G, avoiding division by zero
df['per_user_traffic_LTE'] = df.apply(lambda row: row[spectrum_capacity_column] / row[num_points_column] if row[num_points_column] > 0 else 0, axis=1)
df['per_user_traffic_5G'] = df.apply(lambda row: (row[spectrum_capacity_column] + row[small_cells_capacity_column]) / row[num_points_column] if row[num_points_column] > 0 else 0, axis=1)

# Group by 'geotype' and calculate the maximum average traffic level per user for LTE and 5G
grouped_traffic = df.groupby(geotype_column).agg({
    'per_user_traffic_LTE': 'mean',
    'per_user_traffic_5G': 'mean'
}).reset_index()

# Plot the maximum average traffic level per user for LTE and 5G by geotype
fig, ax = plt.subplots(figsize=(12, 7))

# Set the width of the bars
bar_width = 0.35

# Set the positions of the bars on the x-axis
r1 = np.arange(len(grouped_traffic))
r2 = [x + bar_width for x in r1]

# Make the bars for the LTE and 5G
ax.bar(r1, grouped_traffic['per_user_traffic_LTE'], color='blue', width=bar_width, label='LTE')
ax.bar(r2, grouped_traffic['per_user_traffic_5G'], color='orange', width=bar_width, label='5G')

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Geotype')
ax.set_ylabel('Average Traffic per User (Mbps)')
ax.set_title('Average Traffic per User by Geotype for LTE vs 5G')
ax.set_xticks([r + bar_width / 2 for r in r1])
ax.set_xticklabels(grouped_traffic['geotype'], rotation=45)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('5G_Infrastructure/demand_driven_postcode_data_results.csv')

# Detect relevant columns by keywords
cost_column = next((col for col in df.columns if 'cost' in col.lower()), None)
num_points_column = next((col for col in df.columns if 'numpoints' in col.lower()), None)
geotype_column = next((col for col in df.columns if 'geotype' in col.lower()), None)

# Ensure all necessary columns are found
assert all([cost_column, num_points_column, geotype_column]), "Required columns not found."

# Convert relevant columns to numeric, replacing non-numeric values with NaN
for column in [cost_column, num_points_column]:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Fill NaN values that result from non-numeric data or division by zero with zeros
df.fillna(0, inplace=True)

# Ensure there are no rows with NUMPOINTS equal to zero to avoid division by zero
df = df[df[num_points_column] > 0]

# Calculate investment cost per user
df['Investment.cost.per.user'] = df[cost_column] / df[num_points_column]

# Group by geotype and calculate the mean investment cost per user for each geotype
investment_cost_by_geotype = df.groupby(geotype_column).agg({
    'Investment.cost.per.user': 'mean'
}).reset_index()

# Define colors for each geotype
colors = plt.cm.viridis(np.linspace(0, 1, len(investment_cost_by_geotype)))

# Plot the mean investment cost per user by geotype
plt.figure(figsize=(10, 6))
for i, row in investment_cost_by_geotype.iterrows():
    plt.bar(row[geotype_column], row['Investment.cost.per.user'], color=colors[i], label=str(row[geotype_column]))

plt.title('Average Investment Cost per User by Geotype')
plt.xlabel('Geotype')
plt.ylabel('Investment Cost per User (Currency)')
plt.xticks(rotation=45)
plt.legend(title='Geotype', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

"instruction": "Write a Python script to analyze jitter data in a 5G LTE network traffic dataset. The script should use pandas for data processing and matplotlib for visualization. It should plot jitter data points for each UE and highlight outliers in a contrasting color. Ensure the script includes a legend distinguishing regular data points from outliers and provides clear axis labels."


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv("5G_data/ue-lte-network-traffic-stats.csv")

# Function to find outliers using the Interquartile Range (IQR) method
def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Identify jitter columns
jitter_columns = [col for col in df.columns if 'Jitter' in col]

# Plotting the Jitter data with a different approach to highlight outliers
plt.figure(figsize=(14, 6))

# Plot all jitter data first and identify outliers
for col in jitter_columns:
    plt.scatter(df.index, df[col], alpha=0.5, s=10, label=f'{col} Jitter')
    outliers = find_outliers_iqr(df, col)
    plt.scatter(outliers.index, outliers[col], color='red', s=50, marker='D', label=f'{col} Outliers')

plt.title('Jitter Analysis with Emphasized Outliers')
plt.xlabel('Time Index')
plt.ylabel('Jitter (ms)')
plt.legend()
plt.grid(True)
plt.show()
```

"instruction": "Create a Python script to identify anomalies in 5G LTE network traffic data. Implement an Interquartile Range (IQR) method to detect anomalies in traffic, jitter, and CQI metrics for each User Equipment. The script should calculate the number of anomalies found in each category and store these counts in a DataFrame."

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("5G_data/ue-lte-network-traffic-stats.csv")

# Function to calculate anomalies using IQR
def calculate_anomalies(data, factor=1.5):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - factor * IQR
    upper_bound = Q3 + factor * IQR
    return data.apply(lambda x: 'Anomaly' if x < lower_bound or x > upper_bound else 'Normal')

# Dynamically identify columns related to traffic, jitter, and CQI
traffic_columns = [col for col in df.columns if 'web-rtc' in col or 'sipp' in col or 'web-server' in col]
jitter_columns = [col for col in df.columns if 'Jitter' in col]
cqi_columns = [col for col in df.columns if 'CQI' in col]

# Apply the function to each category of metrics
for col in traffic_columns:
    df[f'{col}_Anomaly'] = calculate_anomalies(df[col])

for col in jitter_columns:
    df[f'{col}_Anomaly'] = calculate_anomalies(df[col])

for col in cqi_columns:
    df[f'{col}_Anomaly'] = calculate_anomalies(df[col])

# Store anomalies in a separate DataFrame and count them
anomaly_columns = [col for col in df.columns if 'Anomaly' in col]
anomalies_df = df[anomaly_columns]
anomaly_counts = anomalies_df.apply(lambda x: x.value_counts().get('Anomaly', 0))

# Printing the counts
print(anomaly_counts)
```

"instruction": "Create a Python script using pandas and matplotlib to calculate and visualize Quality of Service (QoS) metrics for a 5G LTE network traffic dataset. Focus on analyzing jitter, CQI, and throughput for each User Equipment. Calculate the average and standard deviation for these metrics, and plot them in bar charts with error bars representing the standard deviation."

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("5G_data/ue-lte-network-traffic-stats.csv")

# Define the QoS metrics to analyze
jitter_columns = [col for col in df.columns if 'Jitter' in col]
cqi_columns = [col for col in df.columns if 'CQI' in col]
throughput_columns = [col for col in df.columns if 'web-rtc' in col or 'sipp' in col or 'web-server' in col]

# Function to calculate average and standard deviation for a list of columns
def calculate_avg_std(df, columns):
    avg_std = {}
    for col in columns:
        avg_std[col] = {
            'Average': df[col].mean(),
            'Std_Dev': df[col].std()
        }
    return avg_std

# Calculate average and standard deviation for jitter, CQI, and throughput
avg_std_jitter = calculate_avg_std(df, jitter_columns)
avg_std_cqi = calculate_avg_std(df, cqi_columns)
avg_std_throughput = calculate_avg_std(df, throughput_columns)

# Plotting the results
def plot_metrics(avg_std_dict, title, ylabel):
    fig, ax = plt.subplots(figsize=(10, 6))
    metrics = avg_std_dict.keys()
    averages = [avg_std_dict[metric]['Average'] for metric in metrics]
    errors = [avg_std_dict[metric]['Std_Dev'] for metric in metrics]

    ax.bar(metrics, averages, yerr=errors, align='center', alpha=0.7, ecolor='black', capsize=10)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.yaxis.grid(True)

    plt.tight_layout()
    plt.show()

# Plotting Jitter, CQI, and Throughput metrics
plot_metrics(avg_std_jitter, 'Average Jitter with Standard Deviation', 'Jitter (ms)')
plot_metrics(avg_std_cqi, 'Average CQI with Standard Deviation', 'CQI')
plot_metrics(avg_std_throughput, 'Average Throughput with Standard Deviation', 'Throughput')
```
